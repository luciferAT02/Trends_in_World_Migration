```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

file_name="/content/analysis_destination.xlsx"
sheet="migrant_world"
df=pd.read_excel(io=file_name,sheet_name=sheet,index_col="Year")
colors = plt.cm.nipy_spectral(np.linspace(0, 1, 8))
ax=df.plot(color=colors,grid=True,title="Trendline of countries with maximum inflow \n wrt world migrants(both sexes)")
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_ylabel("Migrant stocks as % of total world migrants")
fig=ax.get_figure()
fig.savefig("totalworldmig(1).png")
```
